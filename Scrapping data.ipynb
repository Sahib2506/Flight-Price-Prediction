```python
Price = []
Name = []
Departure= []
Source = []
Duration = []
Stoppages = []
Arrival = []
Destination = []

#Name
n = driver.find_elements_by_xpath("//span[@class='txt-r4 ng-binding']")

for i in n:
    Name.append(i.text)
    
#Departure_Time
dep = driver.find_elements_by_xpath("//span[@class ='txt-r2-n ng-binding']")

for i in range(0,len(dep),2):
    Departure.append(dep[i].text)

#For Source
sr = driver.find_elements_by_xpath("//span[@class ='txt-r3-n ng-binding']")

for i in range(0,len(sr),2):
    Source.append(sr[i].text)


#For Duration
dr = driver.find_elements_by_xpath("//span[@class ='dura_md ng-binding']")

for i in dr:
    Duration.append(i.text)
    
#For Stoppages
st = driver.find_elements_by_xpath("//span[@class ='dura_md2 ng-scope']")

for i in st:
    Stoppages.append(i.text)
    
    
#For Arrival Time

for i in range(1,len(dep),2):
    Arrival.append(dep[i].text)
    
#For Destination

for i in range(1,len(sr),2):
    Destination.append(sr[i].text)
    
#for Price

pr = driver.find_elements_by_xpath("//div[@class ='col-md-8 col-sm-8 col-xs-9 txt-r6-n ng-binding']")

for i in pr:
    Price.append(i.text)
```
